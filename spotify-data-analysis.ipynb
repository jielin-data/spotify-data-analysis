## Section 5: Data Science üßë‚Äçüî¨
The problem I decided to focus on was analyzing the difference in popularity between songs in the 40-50 and 75-85 energy ranges. I chose this because I think it‚Äôs interesting to see how the energy of a song might affect how popular it is. To start, I‚Äôll use Python to split the songs into two groups based on their energy levels. One group will have songs with energy levels between 40 and 50, and the other group will have songs with energy levels between 75 and 85. After that, I‚Äôll perform a z-test(hypothesis test) to check if the difference in popularity between the two groups is statistically significant. I picked this problem because it has clear data, and the smaller energy ranges make it easier to analyze and understand.


## Part 6: Song Counts by Energy Ranges (40‚Äì50 & 75‚Äì85)

#### ‚û°Ô∏è Set the varbile call **‚Äúlow_energy‚Äù** and **‚Äúhigh_energy‚Äù** to find out the answer

```python
import pandas as pd
Spotify=pd.read_csv('Dataset.csv', sep=";")
low_energy = Spotify[(Spotify["energy"] >= 40) & (Spotify["energy"] <= 50)] 
low_energy
```

This code is used there are 191 songs with energy score equal to or less than 40 and less than or equal to 50.

```python
import pandas as pd
Spotify=pd.read_csv('Dataset.csv', sep=";")
high_energy=high_energy = Spotify[(Spotify["energy"] >= 75) & (Spotify["energy"] <= 85)] 
high_energy
```

This code is used there are 580 songs with energy score equal to or less than 75 and less than or equal to 85.

## Part 7ÔºöTotal Songs in Selected Energy Ranges

### Find the total number of songs with energy scores falling in either the 40‚Äì50 or 75‚Äì85 range

### ‚û°Ô∏èTo create a DataFrame containing low and high energy scores and create the varible call "df_score"

```python
df_score = pd.DataFrame({
    "40-50 energy song": low_energy["popularity"],
    "75-85 energy song": high_energy["popularity"],
})
df_score
```

This code shows that there are 771 songs in total with energy scores between 40‚Äì50 and 75‚Äì85.

## Part 8ÔºöDefining Our Hypotheses
### ‚û°Ô∏èDefine the null and alternative hypotheses below. 


The null hypothesis states that there is no difference in the mean popularity between low-energy and high-energy songs, the mean **(H0):low-energy songs = high-energy songs**. The alternative hypothesis is that low-energy songs are more popular on average than high-energy songs,so it is **(H1):low-energy songs > high-energy songs**.

## Part 9:
### Find the average popularity of low energy and high energy

#### ‚û°Ô∏è Set up the varible call **low_mean** and **high_mean** and find out their average popularity

```python
low_mean=low_energy["popularity"].mean()
low_mean
```

This code is find out what is the average popularity of low energy song.

```python
high_mean=high_energy["popularity"].mean()
high_mean
```

This code is find out what is the average popularity of high energy song.

## Part 10ÔºöConducting the Hypothesis Test
### Find out z-test and p-value between energy score 40-50 and 75-85.

```python
from statsmodels.stats.weightstats import ztest
test=ztest(low_energy['popularity'], high_energy['popularity'], alternative='larger')
test
```

This code is find out the z-test and p-value. Also, find out the  Z-test is 2.769 and P- value is smaller than 0.05, so we  reject the null hypothesisÔºålow-energy songs are significantly more popular than high-energy songs (p < 0.05). The large Z-score and tiny p-value confirm this difference is statistically reliable.
